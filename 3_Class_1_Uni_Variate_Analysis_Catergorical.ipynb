```python
import pandas as pd
df = pd.read_csv ("ds.csv")
df.head()
```

```python
# prompt: convert Department to numerical

df['Department'] = pd.factorize(df['Department'])[0]
df.head()
```

```python
a=df["Department"]
```

```python
a.shape
```

```python
# prompt: give count of 1 of a

# Get the count of occurrences of the value 1 in the Series 'a'
count_of_ones = (a == 1).sum()

# Print the count
count_of_ones
```

Basic Measures of Central Tendency

```python
import statistics

# Calculating mean, median, and mode

mode_value = statistics.mode(a)



print("Mode:", mode_value)
```

```python
# prompt: generate bar chart

import matplotlib.pyplot as plt
import numpy as np

# Assuming 'a' is the Series you want to plot
# Calculate the value counts of the Series 'a'
value_counts = a.value_counts()

# Get the unique values and their counts
unique_values = value_counts.index.astype(str) # Convert to string for categorical x-axis
counts = value_counts.values

# Create the bar chart
plt.figure(figsize=(10, 6)) # Optional: Adjust figure size
plt.bar(unique_values, counts)

# Add labels and title
plt.xlabel('Department (Encoded)')
plt.ylabel('Count')
plt.title('Distribution of Departments')

# Show the plot
plt.show()
```

```python
# prompt: generate pie chart of a

# Assuming 'a' is the Series you want to plot
# Calculate the value counts of the Series 'a'
value_counts = a.value_counts()

# Create the pie chart
plt.figure(figsize=(8, 8)) # Optional: Adjust figure size
plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=140)

# Add title
plt.title('Distribution of Departments')

# Ensure the pie chart is a circle
plt.axis('equal')

# Show the plot
plt.show()
```
