```python
import pandas as pd
df = pd.read_csv ("ds.csv")
df.head()
```

```python
a=df["PerformanceRating"]
```

```python
a.shape
```

Basic Measures of Central Tendency

```python
import statistics

# Calculating mean, median, and mode
mean_value = statistics.mean(a)
median_value = statistics.median(a)
mode_value = statistics.mode(a)
std_dev = statistics.stdev(a)  # Sample standard deviation


# Displaying results
print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
print("Standard Deviation:", std_dev)
```

```python
import numpy as np

# Convert "a" to a numpy array
arr = np.array(a)

# Minimum and maximum
min_val = np.min(arr)
max_val = np.max(arr)

# Quartiles
q1 = np.percentile(arr, 25)
q2 = np.percentile(arr, 50)  # Median
q3 = np.percentile(arr, 75)

# Print results
print("Minimum:", min_val)
print("Q1 (25th percentile):", q1)
print("Median (Q2):", q2)
print("Q3 (75th percentile):", q3)
print("Maximum:", max_val)
```

```python
# Calculate specific percentiles, for example 10th, 25th, 50th, 75th, 90th
percentiles_to_get = [10, 25, 50, 75, 90]

# Calculate percentile values
percentile_values = np.percentile(a, percentiles_to_get)

# Print the results
for p, val in zip(percentiles_to_get, percentile_values):
    print(f"{p}th percentile: {val}")
```

```python
# prompt: generate box plot

import matplotlib.pyplot as plt
plt.boxplot(a)
plt.title('Box Plot of Salary')
plt.ylabel('Salary')
plt.show()
```

```python
# prompt: generate histogram 1to 2 , 2 to 3, 3 to4 and 4 to 5 as bins

bins = [1, 2, 3, 4, 5]

plt.hist(a, bins=bins, edgecolor='black')
plt.title('Histogram of Performance Rating with Custom Bins')
plt.xlabel('Performance Rating')
plt.ylabel('Frequency')
plt.xticks(bins) # Set the x-axis ticks to the bin edges
plt.show()
```

```python
# prompt: generate skewness

# Calculating skewness
skewness_value = a.skew()

# Displaying skewness
print("Skewness:", skewness_value)
```

```python
# prompt: explain skewness of -0.2

A skewness of -0.2 indicates that the distribution of your "PerformanceRating" data is slightly skewed to the left.

Here's what that means:

*   **Slightly Skewed:** The value -0.2 is close to 0, which represents a perfectly symmetrical distribution. So, the skewness is not very pronounced.
*   **Left Skew (Negative Skew):** The tail of the distribution is longer on the left side than on the right side. This suggests that there are a few lower values that are pulling the mean down relative to the median.

In simpler terms, while your data is mostly symmetrical, there might be a small number of employees with slightly lower performance ratings that are contributing to this negative skew. The majority of the data is likely clustered towards the higher performance ratings.
```

```python
# prompt: generate a pie chart

# Count the occurrences of each performance rating
rating_counts = a.value_counts()

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Performance Ratings')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# prompt: generate line chart based on histogram

# To generate a line chart based on the histogram, we can use the information from the histogram bins.
# We can get the counts and bin edges from the hist function.
counts, bin_edges = np.histogram(a, bins=20)

# To plot a line chart, we need the x-coordinates (the center of each bin)
# and the y-coordinates (the counts in each bin).
bin_centers = 0.5 * (bin_edges[1:] + bin_edges[:-1])

# Plot the line chart
plt.figure()
plt.plot(bin_centers, counts, marker='o', linestyle='-', color='skyblue')
plt.title('Line Chart (Frequency Polygon) of Salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
