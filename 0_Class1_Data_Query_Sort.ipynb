```python
import pandas as pd
df = pd.read_csv ("ds.csv")
df.head()
```

Filtered Data Frames

```python


# Select only 'Salary' and 'Tier' columns
df2 = df[['Salary', 'Tier']]
df2.head()
```

Conditional Filtering

```python
# Select rows where Salary More > than  100,000, and show only 'Salary' and 'Tier'
df3 = df[df['Salary'] > 30][['Salary', 'Tier']]
df3.head()
```

Filter within a Column

```python
# Get rows where 'Tier' is 'Tier1'
df4 = df[df['Tier'] == 'Tier 1']
df4.head()
```
